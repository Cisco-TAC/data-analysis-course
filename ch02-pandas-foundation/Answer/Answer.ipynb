## 数据选择

### 1. 只选取 ```Goals``` 这一列

````python
euro12.Goals
````

````python
euro12['Goals']
````

### 2. 将列Team, Yellow Cards和Red Cards单独存为一个名叫discipline的数据框

````python
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
discipline
````

````python
discipline = euro12.reindex(columns=['Team', 'Yellow Cards', 'Red Cards'])
discipline
````

### 3. 选取前8行的前7列

````python
euro12.iloc[:8, :7]
````

### 4. 找到进球数Goals超过6的球队数据

````python
euro12[euro12.Goals > 6]
````

### 5. 选取以字母G开头的球队数据

````python
euro12[euro12.Team.str.startswith('G')]
````

## 排序

### 对数据框discipline按照先Red Cards再Yellow Cards进行排序

````python
discipline.sort_values('Yellow Cards').sort_values('Red Cards')
````

````python
discipline.sort_values(['Red Cards', 'Yellow Cards'])
````

## Index操作

### 1. 将队伍名称（Team）设为数据框`discipline`的Index

````python
discipline.index = discipline.Team
discipline
````

````python
discipline.set_index('Team')
````

### 2. 找到英格兰(England)、意大利(Italy)和俄罗斯(Russia)的红黄牌数据

````python
discipline.loc[['England', 'Italy', 'Russia']]
````

````python
discipline[discipline.Team.isin(['England', 'Italy', 'Russia'])]
````

### 3. 射门数最多的球队是哪个？

````python
euro12.Team[euro12.Goals.argmax()]
````

````python
indexes = euro12.index[euro12.Goals.argmax()]
euro12.loc[indexes, ['Team', 'Goals']]
````

````python
euro12.loc[euro12.Goals == euro12.Goals.max(), ['Team', 'Goals']]
````
