## 数据选择

### 列选取

```python
euro12.Goals
```

```python
euro12['Goals']
```

```python
euro12[['Headed goals', 'Goals', 'Passes']]
```

### 行选取

```python
euro12_copy['Croatia':'England']
```

```python
euro12_copy['Croatia':'England'][['Goals', 'Passes']]
```

普通选取的两个问题：
* 行选取必须是范围
* 链式选取（chained indexing）**有可能**返回一个拷贝，也可能返回一个视图

```python
euro12_copy['Croatia':'England']['Goals'] = [1,2,3,4]
euro12_copy['Croatia':'England']['Goals']
```

### 标签（label-based）选取



```python
euro12_copy.loc[['England', 'Denmark', 'Croatia'], ['Goals', 'Passes']]
```

```python
euro12_copy.loc[['England', 'Denmark', 'Croatia', 'Czech Republic'], 'Passes'] = [10,20,30,40]
euro12_copy.loc[['England', 'Denmark', 'Croatia', 'Czech Republic'], ['Goals', 'Passes']]
```

### 下标选取

```python
euro12_copy.iloc[3:7, 2:6]
```

### 混合选取

```python
column_names = euro12_copy.columns[2:6]
euro12_copy.loc[('England', 'France'), column_names]
```

```python
indexes = euro12_copy.index.get_indexer(['England', 'France'])
euro12_copy.iloc[indexes, 2:6]
```

### 条件选取

```python
conditions = euro12_copy.Goals > 5
conditions
```

```python
euro12_copy[conditions]
```

```python
euro12_copy.loc[conditions]
```

```python
euro12_copy.iloc[conditions]
```
