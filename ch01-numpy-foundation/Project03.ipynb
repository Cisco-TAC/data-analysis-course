```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
```

```python
def view_image(image):
    plt.figure()
    if image.ndim == 2:
        plt.imshow(image, cmap=plt.cm.gray)
    else:
        plt.imshow(image)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
image = data.coffee()
view_image(image)
image.shape
```
