```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
```

```python
def view_image(image):
    plt.figure()
    if image.ndim == 2:
        plt.imshow(image, cmap=plt.cm.gray)
    else:
        plt.imshow(image)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
image = data.coffee()
view_image(image)
image.shape
```

### 1. 切割图像——取出图像的右下四分之一

![bottom_right](https://raw.githubusercontent.com/genzj/data-analysis-course/main/data/ch-01/bottom_right.png)

提示：
* 图像大小是多少像素长多少像素宽？（shape）
* 右下四分之一是图像坐标的哪一部分？（`200~400`，`300~600`）
* 切割即是做切片操作，取出需要的坐标部分。（取`[200~400, 300~600]`的部分）

```python
bottom_right = image[200:, 300:, :]
```

```python
view_image(bottom_right)
```

### 2. 缩小图像：降采样法将图像缩小为原图的四分之一

![small](https://raw.githubusercontent.com/genzj/data-analysis-course/main/data/ch-01/small.png)

提示：

* 降采样即每n个点取1个点。
* 缩小为原图的$\frac{1}{4}$，即每4个点取1个点
* 缩小是切片操作，`[0~400每4个元素取一个，0~600每4个元素取一个]`

```python
small = image[::4, ::4, :]
```

```python
view_image(small)
```

### 3. 灰度化：将图像转灰度

![grey](https://raw.githubusercontent.com/genzj/data-analysis-course/main/data/ch-01/grey.png)

提示：

* 彩色图像转灰度，即把3维数组变为2维
* 变化前的数组shape是(400, 600, 3)，变化后是(400, 600)
* 在色彩通道这一维度（第2维）取平均值（`np.mean`）即可把彩色转为黑白

```python
grey = np.mean(image, axis=2)
```

```python
view_image(grey)
```

### 4. 黑白化：将灰度图像转为黑白图像

![bw](https://raw.githubusercontent.com/genzj/data-analysis-course/main/data/ch-01/bw.png)

提示：

* 黑白图像是一个布尔数组
* 灰度图像的shape是(400, 600)，黑白图像的shape不变
* 将灰度图像中的像素与一个阈值（标量）作比较，大于阈值的点为白色（True），小于阈值的点为黑色（False），就得到了黑白图
* 阈值不妨取整个灰度图像的均值，即`np.mean(grey)`
* 黑白图像就是对灰度图像做一个矢量化的比较操作（和`np.mean(grey)`作比较）所得到的布尔值数组。

```python
bw = grey > grey.mean()
```

```python
view_image(bw)
```
