## 蒙特卡罗法求圆周率$\pi$

image.png



1. 二维平面上有一个边长$l=1$的正方形，正方形内有一个内接圆形（即半径$r = 0.5$, 圆心坐标为$(0.5, 0.5)$)
2. 在平面上完全随机地打出N个点
3. 计算有多少点在该圆内（即有多少点坐标$(x, y)$满足$\sqrt{(x-0.5)^2 + (y-0.5)^2} < 0.5$，设结果为$C$
4. 圆内点数与总点数的比值$C : N$和圆与正方形面积的比值$S_{circle} : S_{square}$成正比，即：

  $$\frac{C}{N} = \frac{S_{circle}}{S_{square}} = \frac{\pi \times r ^ 2}{l ^ 2} = \frac{\pi \times 0.5 ^ 2}{1 ^ 2} = \frac{\pi}{4}$$
  
  可得
  
  $$\pi = \frac{C}{N} \times 4 $$
